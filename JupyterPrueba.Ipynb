```python
import matplotlib.pyplot as plt
import numpy as np

# Configurar la fuente globalmente
plt.rcParams['font.family'] = 'DejaVu Serif'

np.random.seed(0)
x = np.random.rand(3000)
y = np.random.rand(3000)

# Generar una variable de grupo (0, 1, 2) para asignar colores diferentes
groups = np.random.choice([0, 1, 2], size=3000)

# Crear una lista de colores
colors = ['red', 'green', 'blue']

# Crear la grfica de dispersin
plt.figure(figsize=(10, 6))
for group in np.unique(groups):
    plt.scatter(x[groups == group], y[groups == group],
                color=colors[group], label=f'Grupo {group}', alpha=0.5, edgecolors='w', s=32)

plt.title('Grfica de dispersin con tres variables')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend(title='Categoras')
plt.grid(True)
plt.show()
```
